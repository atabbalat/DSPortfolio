```python
# Imports
import glob
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions
from tensorflow.keras.applications import resnet50
```

## Importing Data

```python
model = ResNet50(weights = 'imagenet')
```

## Classification

```python
# Define a function for classifying an image
def classify_image(img_path):
  # Load the image and resize it to 224x224
  img = image.load_img(img_path, target_size=(224, 224))

  # Visualize the image
  plt.imshow(img)
  plt.show()

  # Convert the image to a numpy array, preprocess it, and add a batch dimension
  image_array = image.img_to_array(img)
  image_array = np.expand_dims(image_array, axis=0)
  image_array = preprocess_input(image_array)

  # Use the trained model to predict the class probabilities
  preds = model.predict(image_array)

  # Write the prediction to a file
  with open('results/ResNet50/results.txt', 'a') as f:
    f.writelines(str(preds))

  # Print the top 3 predicted classes and their probabilities
  print('Predicted:', decode_predictions(preds, top=3)[0])
```

```python
# Bear image
classify_image('images/bear.png')
```

```python
# Elephant image
classify_image('images/elephant.png')
```
